```python
from ultralytics import YOLO

import matplotlib.pyplot as plt

import cv2
```

```python
model_path = 'yolo26n.pt'

model = YOLO(model_path)

#path_img_test = 'autoriaNumberplateOcrRu\\test\\img\\A003CX196.png'
path_img_test = 'data_car_number_labels\\images\\train\\408.jpg'

image_test = cv2.imread(path_img_test)
image_test = cv2.cvtColor(image_test, cv2.COLOR_BGR2RGB)

plt.imshow(image_test)
plt.show()

results = model(path_img_test)
```

```python
boxes = results[0].boxes

for idx, box in enumerate(boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
    x1 = max(0, x1)
    y1 = max(0, y1)
    x2 = min(image_test.shape[1], x2)
    y2 = min(image_test.shape[1], y2)

    print(box.xyxy[0])

    crop = image_test[y1:y2, x1:x2]
    plt.imshow(crop)
    plt.show()
```
